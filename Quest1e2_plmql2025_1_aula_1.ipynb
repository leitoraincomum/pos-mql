```python
# Carregando uma biblioteca
import pandas as pd
```

```python
#importar um arquivo
from google.colab import files
uploaded = files.upload()
```

```python
# transformando o arquivo em conjunto de dados
import io
df = pd.read_excel(io.BytesIO(uploaded['Base de dados a limpar.xlsx']))
```

```python
df.info()
```

```python
# contar valores ausentes
pd.isna(df).sum()
```

```python
# Identificando observações vazias pela variável mais relevante
pd.isnull(df['Entrega']).sum()
```

```python
# Removendo as observações vazias
df = df.dropna(how='all')
```

```python
df
```

```python
# contar valores ausentes
pd.isna(df).sum()
```

```python
# Retirar observações com valor ausente na variável Cliente
df = df.dropna(subset=['cliente'])
```

```python
# Contar valores ausentes
pd.isna(df).sum()
```

```python
df
```

```python
df["Km_percorrida"]=df["Km_percorrida"].fillna(df["Km_percorrida"].mean())
```

```python
# Contar valores ausentes
pd.isna(df).sum()
```

```python
# Identificar observações duplicadas
df.duplicated().sum()
```

```python
# excluindo observações duplicadas
df = df.drop_duplicates()
```

```python
df
```

```python
# Verificar se há erros de grafia em variáveis categóricas
df["Cidade"].unique()
```

```python
# Corrigir a grafia de Mogi das Cruzes
df["Cidade"]=df["Cidade"].replace({"Mogi dasCruzes":'Mogi das Cruzes'},regex=True)
```

```python
# Verificar se há erros de grafia em variáveis categóricas
df["Cidade"].unique()
```

```python
# Corrigir a grafia de Suzno e Poa (multiplas correções)
df["Cidade"]=df["Cidade"].replace({"Suzno":'Suzano',"Poa":'Poá'},regex=True)
```

```python
# Verificar se há erros de grafia em variáveis categóricas
df["Cidade"].unique()
```

```python
# Exportando o arquivo tratado
from google.colab import files
df.to_excel("df.xlsx")
files.download("df.xlsx")
```
